```python
import sagemaker
from sagemaker.xgboost.model import XGBoostModel
from sagemaker import get_execution_role
```

```python
session = sagemaker.Session()
role = get_execution_role()

model_path = "s3://sagemaker-churn-mlops-ka/sagemaker-churn-mlops/model/churn-xgb-train-2025-12-02-01-41-14-611/output/model.tar.gz"
```

```python
xgb_model = XGBoostModel(
    model_data=model_path,
    role=role,
    entry_point=None,  # No custom inference logic needed for now
    framework_version="1.2-1",
    sagemaker_session=session
)
```

```python
predictor = xgb_model.deploy(
    initial_instance_count=1,
    instance_type="ml.t2.medium",
    endpoint_name="churn-xgb-endpoint"
)

print("âœ… Model deployed and live at endpoint: churn-xgb-endpoint")
```

```python
from sagemaker.serializers import CSVSerializer
predictor.serializer = CSVSerializer()

# 15 features in correct order (no header)
sample_payload = "41,28,28,7,13,584,20,0,1,0,1,0,0,1,0"  # matches training order

result = predictor.predict(sample_payload)
print("ðŸ”® Prediction:", result)
```
