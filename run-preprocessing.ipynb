```python
import sagemaker
from sagemaker.sklearn.processing import SKLearnProcessor
from sagemaker.processing import ProcessingInput, ProcessingOutput
import boto3
import os
```

```python
session=sagemaker.Session()
role=sagemaker.get_execution_role()
bucket='sagemaker-churn-mlops-ka'
```

```python
sklearn_processor = SKLearnProcessor(
    framework_version='1.2-1',
    role=role,
    instance_type='ml.t3.medium',
    instance_count=1,
    base_job_name='churn-preprocessing',
    sagemaker_session=session
)
```

```python
sklearn_processor.run(
    code='processing/processing.py',
    inputs=[
        ProcessingInput(
            source='s3://sagemaker-churn-mlops-ka/sagemaker-churn-mlops/raw/churn.csv',
            destination='/opt/ml/processing/input'
        )
    ],
    outputs=[
        ProcessingOutput(
            source='/opt/ml/processing/train',
            destination='s3://sagemaker-churn-mlops-ka/sagemaker-churn-mlops/processed/train'
        ),
        ProcessingOutput(
            source='/opt/ml/processing/test',
            destination='s3://sagemaker-churn-mlops-ka/sagemaker-churn-mlops/processed/test'
        )
    ]
)
```
